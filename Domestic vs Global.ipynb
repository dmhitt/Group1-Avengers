```python
#import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#turn csv containing monetary information into a dataframe 
marvel_movies_df= pd.read_csv('mcu_domestic_global_box_office.csv')
marvel_movies_df
```

```python
#create data frame that has movies featuring Chris Evans as Captain America and Robert Downey Jr. as Iron Man
iron_vs_cap_df= pd.read_csv('rdj_ce_movie_data.csv')
iron_vs_cap_df
```

```python
#merge both data frames
merged_df=marvel_movies_df.merge(iron_vs_cap_df, how='left')
merged_df=merged_df.dropna()


merged_df
```

```python
actor_count= merged_df.set_index(["Actor", "Movie", "Release Date", "Opening Weekend"]).count(level="Actor")
print(actor_count)
```

```python
#turn string values into numeric values
merged_df['Domestic Box Office']= merged_df['Domestic Box Office'].replace('[\$\,\.]',"",regex=True).astype(float)
merged_df['Worldwide Box Office']= merged_df['Worldwide Box Office'].replace('[\$\,\.]',"",regex=True).astype(float)
merged_df['Production Budget']= merged_df['Production Budget'].replace('[\$\,\.]',"",regex=True).astype(float)
```

```python
#plot domestic box office for each movie
domestic_df=merged_df
domestic_df=domestic_df.drop_duplicates(subset='Movie', keep='first')
domestic_df.plot(kind='bar', x='Movie', y=['Domestic Box Office', 'Production Budget'], figsize=(15,15), color=['r','y'])
plt.title('Domestic Box Office per Movie',fontsize=20, fontweight='bold')
plt.xticks(rotation=80)
plt.ylabel('Dollars (Hundred Millions)', fontsize=18, fontweight='bold')
plt.xlabel('Movie', fontsize=18, fontweight='bold')
plt.show()
```

```python
#plot worldwide box office for each movie
global_df= domestic_df
global_df.plot(kind='bar', x='Movie', y= ['Worldwide Box Office', 'Production Budget'], figsize=(15,15), color=['b', 'y'])
plt.title('Worldwide Box Office per Movie',fontsize=20, fontweight='bold')
plt.xticks(rotation=80)
plt.ylabel('Dollars (Billions)', fontsize=18, fontweight='bold')
plt.xlabel('Movie', fontsize=18, fontweight='bold')
plt.show()
```

```python
#combine domestic and worldwide plots. Include production budget. 
joint_box_office= global_df
joint_box_office.plot(kind='bar', x='Movie', y= ['Worldwide Box Office', 'Domestic Box Office', 'Production Budget'], 
                      figsize=(20,15), color=['r', 'b', 'y'], width=0.8)
plt.title('Worldwide Box Office per Movie',fontsize=20, fontweight='bold')
plt.xticks(rotation=80)
plt.ylabel('Dollars (Billions)', fontsize=18, fontweight='bold')
plt.xlabel('Movie', fontsize=18, fontweight='bold')
plt.show()
```

```python
#Create profit
domestic_production= merged_df.drop(['Release Date', 'Opening Weekend', 'Worldwide Box Office'], axis=1 )
domestic_production['Profit']=domestic_production['Domestic Box Office']-domestic_production['Production Budget']

#Find and print total for 
Actor_profit_domestic= domestic_production.groupby(['Actor'])
Actor_profit_domestic_value= Actor_profit_domestic['Profit'].sum()
print(f"The total domestic profit for Chris Evans is {'${:,.0f}'.format(Actor_profit_domestic_value['Chris Evans'])}")
print(f"The total domestic profit for Robert Downey Jr. is {'${:,.0f}'.format(Actor_profit_domestic_value['Robert Downey Jr.'])}")
```

```python
avg_domestic_RDJ= Actor_profit_domestic_value['Robert Downey Jr.']/9
avg_domestic_CE= Actor_profit_domestic_value['Chris Evans']/7
print(f'The profit per movie for Robert Downey Jr. as Iron Man domestically is {"${:,.0f}".format(avg_domestic_RDJ)}')
print(f'The profit per movie for Chris Evans as Captain America domestically is {"${:,.0f}".format(avg_domestic_CE)}')
```

```python
domestic_production['Domestic Box Office']=domestic_production['Domestic Box Office'].apply(lambda x: '${:,.0f}'.format(x))
domestic_production['Profit']=domestic_production['Profit'].apply(lambda x: '${:,.0f}'.format(x))
domestic_production['Production Budget']=domestic_production['Production Budget'].apply(lambda x: '${:,.0f}'.format(x))
domestic_production
```

```python
#Create profit
global_production= merged_df.drop(['Release Date', 'Opening Weekend', 'Domestic Box Office'], axis=1 )
global_production['Profit']=global_production['Worldwide Box Office']-global_production['Production Budget']

#find total
Actor_profit_global= global_production.groupby(['Actor'])
Actor_profit_global_value= Actor_profit_global['Profit'].sum()
print(f"The total worldwide profit for Chris Evans is {'${:,.0f}'.format(Actor_profit_global_value['Chris Evans'])}")
print(f"The total worldwide profit for Robert Downey Jr. is {'${:,.0f}'.format(Actor_profit_global_value['Robert Downey Jr.'])}")
```

```python
avg_global_RDJ= Actor_profit_global_value['Robert Downey Jr.']/9
avg_global_CE=Actor_profit_global_value['Chris Evans']/7
print(f'The profit per movie for Robert Downey Jr. as Iron Man worldwide is {"${:,.0f}".format(avg_global_RDJ)}')
print(f'The profit per movie for Chris Evans as Captain America worldwide is {"${:,.0f}".format(avg_global_CE)}')
```
